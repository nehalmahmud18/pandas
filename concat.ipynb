```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'roll no':[1,2,3,4], 'physics':[34,45,56,78], 'chemistry':[12,23,39,48]})
df2 = pd.DataFrame({'roll no':[5,6,7,8], 'physics':[37,45,50,98], 'chemistry':[90,34,89,84]})

print(df1)
print(df2)
```

```python
df3 = pd.concat([df1,df2])
df3
```

```python
df4 = pd.concat([df1,df2], ignore_index= True)
df4
```

```python
df4 = pd.concat([df1,df2], ignore_index= True, sort= True)
df4
```
