#merge() function

```python
df1 = pd.DataFrame({'Roll No':[1,2,3,4],'Physics':[32,34,56,76]})
df1
```

```python
df2 = pd.DataFrame({'Roll No':[1,2,3,4],'Chemistry':[45,56,67,78]})
df2
```

```python
pd.merge(df1, df2, on = 'Roll No')
```

```python
pd.merge(df2, df1)
```

```python
df3 = pd.DataFrame({'roll no':[2,3,4,5],'physics':[23,34,56,67]})
df4 = pd.DataFrame({'roll no':[2,3,4,9],'Chemistry':[30,34,90,67]})

pd.merge(df3, df4, on= 'roll no')
```

```python
pd.merge(df3,df4, how = 'left')
```

```python
pd.merge(df3,df4, how = 'right')
```

```python
pd.merge(df3,df4, how = 'outer')
```
